```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sheri=pd.DataFrame({

    'sheri2': [20, 30, 40, 50, 70],
    'sheri1': [10, 32, 42, 52, 72]
    
})


```

```python
sheri['sheri2'].plot(kind='line', color='g', label='sheri2', linewidth=1, alpha=0.3,  linestyle='-')
sheri['sheri1'].plot(color='b', label='sheri1', linewidth=1, alpha=0.5, linestyle='-')

# Adding legend, labels, and displaying the plot
plt.legend(loc='upper right')  # legend = puts label into plot
plt.xlabel('x axis')           # label = name of label
plt.ylabel('y axis')           # label = name of label
plt.show()
```
